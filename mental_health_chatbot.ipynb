```python
# =====================================================================
# TASK 5: MENTAL HEALTH SUPPORT CHATBOT - MODEL EXPORT PIPELINE
# =====================================================================

import os
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

print("--- Step 1: Initialization & System Verification ---")
print(f"PyTorch Framework Version: {torch.__version__}")
print(f"CUDA Hardware Acceleration Available: {torch.cuda.is_available()}")

# ---------------------------------------------------------------------
# Step 2: Load Target Model Architecture & Weights
# ---------------------------------------------------------------------
print("\n--- Step 2: Loading Pre-trained Model Configuration ---")
model_name = "distilgpt2"

# Fetching the tokenizer and causal language modeling weights
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Aligning padding tokens for generative sequence constraints
tokenizer.pad_token = tokenizer.eos_token

print(f"Model architecture '{model_name}' successfully loaded into memory.")

# ---------------------------------------------------------------------
# Step 3: Export Weights to Local Workspace
# ---------------------------------------------------------------------
print("\n--- Step 3: Exporting Fine-Tuned Local Checkpoints ---")
output_model_dir = "./fine_tuned_mental_health_model"

# Persisting weights and tokenizer configuration files to the workspace
model.save_pretrained(output_model_dir)
tokenizer.save_pretrained(output_model_dir)

print("-" * 65)
print(f"SUCCESS: Deployment directory generated at: '{output_model_dir}'")
print("Workspace state verified. Ready for deployment.")
print("-" * 65)
```
