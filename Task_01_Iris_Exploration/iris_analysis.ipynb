```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")
df = sns.load_dataset("iris")

print(f"Dataset Shape: {df.shape}")
print(f"columns Available: {list(df.columns)}\n")
df.head()
```

```python
print("--- Dataset Information ---")
df.info()

print("\n--- Summary Statistics ---")
df.describe()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle(
    "Iris Dataset Exploratory Data Analysis", fontsize=12, fontweight="bold"
)

sns.scatterplot(
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette="deep",
    ax=axes[0, 0],
)
axes[0, 0].set_title("Sepal Length vs Width")

sns.scatterplot(
    data=df,
    x="petal_length",
    y="petal_width",
    hue="species",
    palette="deep",
    ax=axes[0, 1],
)
axes[0, 1].set_title("Petal Length vs Width")

sns.histplot(
    data=df,
    x="petal_length",
    hue="species",
    kde=True,
    multiple="stack",
    palette="deep",
    ax=axes[1, 0],
)
axes[1, 0].set_title("Distribution of Petal Length")

sns.boxplot(
    data=df,
    x="species",
    y="sepal_width",
    hue="species",
    palette="deep",
    legend=False,
    ax=axes[1, 1],
)
axes[1, 1].set_title("Sepal Width Outlier Detection")
plt.tight_layout()
plt.show()
```
